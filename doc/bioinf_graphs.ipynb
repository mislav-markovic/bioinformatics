```python
import matplotlib.pyplot as plt
time = [0.1, 1, 15, 170, 1830, 8052, 16208]
length = [100, 1000, 10000, 100000, 1000000, 4647121, 10000000]
memory = [0.4, 5.5, 6, 61, 541, 2448, 4957]
```

```python
print(" ")
plt.plot(time, length, 'bo', time, length)
plt.ylabel("string size")
plt.xlabel("time")
plt.show()
print(" ")
```

```python
print(" ")
plt.plot(memory, length, 'bo', memory, length)
plt.ylabel("string size")
plt.xlabel("memory usage")
plt.show()
print(" ")
```
